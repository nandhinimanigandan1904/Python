```python
#Body Mass Index
#Used chart reference for women  with age 30 to 40 
#Underweight <18.5 ,Normal 18.5 - 24,Overweight 25- 29 ,Very overweight 30 and above 

bmi_value = float(input("Enter the BMI Index:"))
if(bmi_value<18.5):
    print("Underweight")
elif(bmi_value<25):
    print("Normal")
elif(bmi_value<30):
    print("Overweight")
else:
    print("Very Overweight")

```
