```python
# print Odd Numbers in the list
#Expected Output 
#(20,10,16,19,25,1,276,188)
#19 is odd
#25 is odd
#1 is odd

list = [20,10,16,19,25,1,276,188]
print(list)
for val in list:
    if (val%2 == 1):
        print (val,"is odd")
```

```python
# print Even numbers in the list
#expected output
"""
(20,10,16,19,25,1,276,188)
20 is odd
10 is odd
16 is odd
276 is odd
188 is odd 
"""
list = [20,10,16,19,25,1,276,188]
print (list)
for val in list:
    if (val%2 == 0):
        print (val,"is even")



```
