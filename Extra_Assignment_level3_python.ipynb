```python
#Check the password using if and else 
import getpass

# Standard hidden input
passwd = getpass.getpass("Enter the password: ")

if (passwd == "HOPE@123"):
    print("Your password is correct")
else:
    print("Your password is invalid")

          
```

```python
#Category the people by their age like children ,adult ,citizen,senior citizen 
#Children: 0–14 yearsYouth: 15–24 yearsYoung Adults: 25–44 years Citizen Age: 45–59 yearsElderly / Older Adults: 60+ senior citizen 
age = int(input("Enter your Age"))
if (age <0):
    print ("Invalid Age")
elif (age<=1):
    print("Infant")
elif (age<=14):
    print ("Children")
elif (age<=24):
    print ("Youth")
elif (age<=44):
    print("Adults")
elif (age<=59):
    print("Citizen")
else:
    print("Senior Citizen")
```
